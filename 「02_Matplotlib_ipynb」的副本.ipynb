### [Matplotlib](https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 4*np.pi, 0.1) # 0~12.
y = np.sin(x)
print(x[:10])
print(y[:10])

plt.plot(x, y)
plt.show()
```

### Basic

```python
x = np.arange(0, 4*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Sine and Cosine")
plt.legend(["Sine", "Cosine"]) 

plt.show()
```

```python
# Method2: Plotting multiplt lines

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y_sin)
axes.plot(x,y_cos)
axes.set_xlabel("x axis label")
axes.set_ylabel("y axis label")
axes.set_title("Sine and Cosine")
plt.legend(["Sine", "Cosine"])
plt.show()
```

### Subplot

```python
# set figure size
plt.figure(figsize=(20, 3)) # (x, y)

# (row size, column size, index(start from 1))
plt.subplot(1, 2, 1)
plt.plot(x, y_sin)
plt.title("Sine")

plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title("Cosine")

plt.show()
```

### Scatter plot

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y, s=None, c=None, marker=None, alpha=None)
plt.title("Scatter plot")
plt.show()
```

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y, 
            s=100, # size 
            c='green', 
            marker="^", 
            alpha=0.5)
plt.title("Scatter plot")
plt.show()
```

#### Bar

```python
X = np.arange(10)
Y = np.random.normal(0, 1, 10)
plt.bar(X, Y)
plt.show()
```

### Histogram

```python
data = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(data, bins=20)
plt.show()
```

### Plot image

```python
img = np.random.randint(0, 256, (4, 4))

print(img)
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img, cmap='gray') # color map
```

```python
img = np.random.randint(0, 256, (4, 4, 3)) #  (H, W, RGB Channels)
print(img)
plt.imshow(img)
plt.show()
```

```python
img = np.random.randint(0, 256, (32, 32))
plt.imshow(img, cmap='coolwarm')
```
